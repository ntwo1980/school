```python
import io

def outputChar(io, ch):
    if ch == "\n":
        output.write("</br>")
    else:
        output.write(ch)

article = """又一个黎明，很快会到来，到那时，老计又要对一列列火车迎来送往。但，我深信，每一次举起臂膊的时候，老计都是那么幸福，因为，在他的生命中，那时一定“彗星出现”。"""

output = io.StringIO()
article_len = len(article)
# biaodian = {"，", "；", "。", "“", "”", "、", "：", "！", "？", "一"} 
biaodian = {"，", "；", "。", "“", "”", "、", "：", "！", "？"} 
start_num = 15
current_num = start_num

for i in range(article_len):
    ch = article[i]
    
    if ch not in biaodian:
        outputChar(output, '{{c')

        if current_num == start_num:
            outputChar(output, str(current_num))
            current_num += 1
        else:
            outputChar(output, str(current_num))
            current_num -= 1
            
        outputChar(output, '::')
        outputChar(output, ch)
        outputChar(output, '}}')
    else:
        outputChar(output, ch)

print(output.getvalue())
```
