```python
from http import HTTPStatus
import dashscope

dashscope.api_key="sk-2bd35d7b0181490b909da641dbc7eb92"

def call_with_messages():
    messages = [{'role': 'system', 'content': '你是高考语文命题组成员.'},
                {'role': 'user', 'content': '我想背诵内容：一星陨落，黯淡不了星空灿烂；一花凋零，荒芜不了整个春天。请给它打一个分值，代表它对高考作文的价值。分值范围从1到10。你的答案只需要给一个确定的指数值，不需要解释。'}]

    response = dashscope.Generation.call(
        dashscope.Generation.Models.qwen_turbo,
        messages=messages,
        result_format='message',  # 将返回结果格式设置为 message
    )
    if response.status_code == HTTPStatus.OK:
        print(response)
    else:
        print('Request id: %s, Status code: %s, error code: %s, error message: %s' % (
            response.request_id, response.status_code,
            response.code, response.message
        ))

# 示例：向通义千问提问
question = "你好，通义千问，你是谁？"
answer = call_with_messages()
if answer:
    print("回答:", answer)
```
